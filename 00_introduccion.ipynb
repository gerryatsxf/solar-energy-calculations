# Base de datos de irradiancia

El Laboratorio Nacional de Energías Renovables (NREL, por sus siglas en inglés) es un instituto de investigación dedicado al avance de las tecnologías de energías renovables y eficiencia energética. Establecido en 1977 y ubicado en Golden, Colorado, NREL es una división del Departamento de Energía de los Estados Unidos (DOE) y es considerado un importante centro de investigación y desarrollo de energías renovables. La misión de NREL es promover un futuro de energía sostenible mediante la realización de proyectos de investigación y desarrollo, y proporcionando al público la información y las herramientas que necesitan para tomar decisiones informadas sobre energía.

Una de las contribuciones clave de NREL al campo de la energía solar son sus bases de datos de irradiancia, que proporcionan información detallada sobre la cantidad de radiación solar que llega a la superficie terrestre en diferentes lugares y momentos. Estos datos son críticos para el diseño, análisis y optimización de sistemas de paneles solares, y son utilizados por investigadores, ingenieros y fabricantes de todo el mundo.

Para producir datos precisos y fiables, NREL opera una red de estaciones de medición terrestres y satelitales que monitorean los niveles de radiación solar en tiempo real. Los datos recogidos por estas estaciones son analizados y validados por el equipo de expertos de NREL y están disponibles al público. Este laboratorio también desarrolla herramientas avanzadas de modelado y algoritmos que pueden usarse para estimar los niveles de radiación solar en cualquier lugar, basados en datos de sus estaciones de medición y otras fuentes.

Una de las aplicaciones de los datos de irradiancia de NREL es el diseño y optimización de sistemas de paneles solares. Al utilizar información precisa y actualizada sobre la cantidad de radiación solar que llega a la superficie de la Tierra, los ingenieros y fabricantes pueden diseñar paneles solares que estén optimizados para las condiciones climáticas y meteorológicas específicas de una ubicación determinada. Esta información también se puede utilizar para modelar el rendimiento de los paneles solares con el tiempo, permitiendo evaluar los beneficios a plazos de la energía solar.

El NREL es un agente crítico en el campo de la energía renovable, y su base de datos de irradiancia es un recurso valioso para aquellos interesados en la energía solar. Ya sea que sea un investigador, ingeniero o personajes interesados en el tema por cuestiones políticas y normativas, NREL proporciona los datos y las herramientas necesarias para tomar decisiones informadas sobre la energía solar, ya sea que esté interesado en optimizar el rendimiento de un sistema de paneles solares o en conocer los beneficios y oportunidades económicas y financieras en la energía renovable.

### Paneles solares

Los paneles solares pueden generar electricidad gracias al efecto fotoeléctrico. Las celdas solares están compuestas por materiales semiconductores, como silicio, cuyos electrones son excitados por la incidencia de fotones solares y aumenta sus niveles de energía. Los electrones, al alcanzar un cierto nivel de energía, son liberados de las órbitas atómicas de estos materiales semiconductores y generan una diferencia de potencial entre las capaz de estos materiales. Esta diferencia de potencial permite que los electrones fluyan generando una corriente eléctrica. La corriente eléctrica generada es luego enviada a un inversor que la convierte en una forma utilizable de energía, como corriente alterna. La eficiencia de un panel solar depende de la cantidad de luz solar incidente y de la capacidad del material semiconductor para generar y conducir la corriente eléctrica.

La incidencia en el panel solar puede ser modelada como la proyección de irradiancia en un plano generalizado; definimos este plano generalizado como un plano que atraviesa el punto de la superficie terrestre donde se encuentra nuestro observador y que además es perpendicular a un vector normal $\hat{s}$, que poseerá un ángulo con respecto zenith del observador.

Definimos un punto cualquiera de la superficie terreste como la posición de nuestro observador (punto donde se encontrará nuestro sistema de paneles solares), un vector normal a la superficie terrestre $\hat{n}$ y al ángulo $\beta$ como el ángulo de inclinación del plano con respecto al plano tangente a la superficie terrestre y que atraviesa las coordenadas del observador, también conocido como plano del observador. El ángulo de inclinación $\beta$ es equivalente al ángulo entre $\hat{s}$ y $\hat{n}$
